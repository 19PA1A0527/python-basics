```
from matplotlib import pyplot as plt
plt.style.use("fivethirtyeight")
slices=(60,40)
labels=["sixty","eighty"]
plt.pie(slices,labels=labels)
plt.title("my pie chart")
plt.show()
# pie plots are generally used for labels containing less slices(generally barplots are used for more more data)
```

```
from matplotlib import pyplot as plt
plt.style.use("fivethirtyeight")
slices=[59904,55678,47658,36789,35467]
labels=['java','html','sql','python','java']
explode=[0,0,0,0.1,0]
plt.pie(slices,labels=labels,explode=explode)
plt.title("my pie chart")
plt.show()
```
